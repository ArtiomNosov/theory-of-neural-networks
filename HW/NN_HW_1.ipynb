```python
# HW 1.1
'''
program a recurrent neural network and achieve all states:
1) Fixed points;
2) Limit cycle;
3) The strange attractor is not a fixed point;
4) Divergent dynamics.
'''
print()
```

```python
weights = np.array([
    [-2, 1, 0],
    [1, 0.5, 1],
    [2, -2, 0]
])
bias = np.array([1, -1, 2])
s = np.array([1, 0, 1])

def y1(h):
  if h >= 0:
    return h
  return 0

def y2(h):
  return h

def y3(h):
  return y1(h)

functions = [y1, y2, y3]
```

```python
def RNN(x_sequence, weights, bias, s0, functions):
  s = cy.copy(s0)
  result = np.reshape(np.array([0, 0, 0]), (1, 3))
  for x in x_sequence:
    h = np.sum((weights.T * s).T, axis=0) + bias
    s[0] = functions[0](h[0] + x)
    for i in range(1, len(h)):
      s[i] = functions[i](h[i])
    result = np.append(result,np.reshape(s, (1, 3)), axis=0)
    # print(s)
  return result
```

```python
# 1) Fixed points
s = np.array([0, -2, 0])
result = RNN([i % 2 for i in range(0, 100)], weights, bias, s, functions)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(result[1:,0], result[1:,1],  result[1:,2])
print(result[1:5])
```

```python
# 2) Limit cycle.
s = np.array([1, 0, 1])
result = RNN([i % 2 for i in range(0, 100)], weights, bias, s, functions)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(result[1:,0], result[1:,1],  result[1:,2])
print(result[1:13])
```

```python
# 3) The strange attractor is not a fixed point
weights = np.array([
    [-2, 1, 0.5],
    [-1, 0.5, 1],
    [1, -2, -0.5]
])
result = RNN([random.randint(0, 1) for i in range(0, 100)], weights, bias, s, functions)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(result[1:,0], result[1:,1],  result[1:,2])
print(result[1:20])
```

```python
# 4) Divergent dynamics.
weights = np.array([
    [-2, 1, 1],
    [1, 0.5, 0],
    [2, -2, 3]
])
s = np.array([1.0, 0.0, 1.0])
result = RNN([i % 2 for i in range(0, 100)], weights, bias, s, functions)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(result[1:,0], result[1:,1],  result[1:,2])
print(result[1:15])
```
